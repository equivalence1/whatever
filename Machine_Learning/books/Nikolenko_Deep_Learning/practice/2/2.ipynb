```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# Linear regression
```

```python
k, b = np.random.rand(2)
print(k, b)
```

```python
x = np.linspace(0, 10, 100)
y = k * x + b

plt.plot(x, y, label=('k=' + str(k) + ", b=" + str(b)))
plt.show()
```

```python
t = k * x + b + np.random.normal(0, 0.3, len(x))

plt.plot(x, y, '--r', label=('k=' + str(k) + ", b=" + str(b)))
plt.scatter(x, t)
plt.show()
```

```python
def linear_reg_solve(X, t, step=0.01, N=100):
    x = torch.cat((X.view(-1, 1), torch.ones(len(X), dtype=X.dtype).view(-1, 1)), dim=1)
    w_hat = torch.rand((2, 1), requires_grad=True, dtype=X.dtype)
    
    for n in range(N):
        predictions = torch.mm(x, w_hat)

        with torch.no_grad():
            plt.clf()
            plt.scatter(X, t)
            t_hat = torch.mm(x, w_hat)
            plt.plot(X, t_hat, '--r')
            plt.pause(0.1)
            plt.show()
        
        l2 = (t - predictions.view(-1)) ** 2 / len(t)
        print('L2: ', l2.sum())
        print('w_hat: ', w_hat)

        if w_hat.grad is not None:
            print(w_hat.grad)
            w_hat.grad *= 0
        l2.sum().backward()
        w_hat.data.sub_(step * w_hat.grad.data)
        
    return w_hat
```

```python
w_hat = linear_reg_solve(torch.from_numpy(x), torch.from_numpy(t))
```

```python
print('k = ', k, ', b = ', b, ', w_hat = ', w_hat)
```

```python
# Logistic regression
```

```python
X1 = torch.distributions.multivariate_normal.MultivariateNormal(torch.tensor([-.9, -.9]), torch.tensor([[0.3, 0.2], [0.2, 0.3]])).sample((20,))
X2 = torch.distributions.multivariate_normal.MultivariateNormal(torch.tensor([.8, .8]), torch.tensor([[0.3, 0.2], [0.2, .3]])).sample((20,))
```

```python
X = torch.cat((X1, X2), dim=0)
y = [0 if i < len(X1) else 1 for i in range(len(X))]
```

```python
plt.scatter(X1.numpy()[:, 0], X1.numpy()[:, 1], marker='x')
plt.scatter(X2.numpy()[:, 0], X2.numpy()[:, 1], marker='+')
plt.show()
```

```python
def logistic_regression_solve(X1, X2, step=0.01, N=100):
    X = torch.cat((X1, X2), dim=0)
    y = torch.tensor([0. if i < len(X1) else 1. for i in range(len(X))])
    x = torch.cat((X, torch.ones(len(X)).view(-1, 1)), dim=1)
    w_hat = torch.rand((3, 1), requires_grad=True)
    
    for it in range(N):
        with torch.no_grad():
            line_x = x[:, 0]
            line_y = (-line_x * w_hat[0] - w_hat[2]) / (w_hat[1] + 1e-7)
            plt.clf()
            plt.plot(line_x, line_y, '--r')
            plt.scatter(X1.numpy()[:, 0], X1.numpy()[:, 1], marker='x')
            plt.scatter(X2.numpy()[:, 0], X2.numpy()[:, 1], marker='+')
            plt.pause(0.1)
            plt.show()
            
        predictions = torch.nn.functional.sigmoid(torch.mm(x, w_hat)).view(-1)
        ce = -1 * (y * torch.log(predictions) + (1 - y) * torch.log(1 - predictions))
        
        if w_hat.grad is not None:
            w_hat.grad *= 0
        ce.sum().backward()
        w_hat.data.sub_(step * w_hat.grad.data)
        
    return w_hat
```

```python
w_hat = logistic_regression_solve(X1, X2)
```
